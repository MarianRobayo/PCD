```python
#Para utilizar if en la preparación del conunto de datos, se decide clasificar una variable de tipo entero en categorias tipo string para que
#la visualización e interpretación de los datos sea más sencilla.

#definición de la nueva columna que se llamará clase
df['clase']= [ clasificar(x) for x in df['Pclass']]
```

```python
#En esta sección se le asigna cada uno de los valores de la variable a su correspondiente valor en la nueva columna.
def clasificar(valor):
    if valor==1:
        return '1st class'
    elif valor==2:
        return '2nd class'
    else:
        return '3rd class'  
```
