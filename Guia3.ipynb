En el df1 hay 7 variables que tienen valores nulos mientras que en el df2 no hay.

```python
# A continuación revisaremos la variable NIT en cada uno de los dataframes
# Vamos a identificar los NITs que tienen más de 9 digitos, para ello crearemos una nueva columna que almacenara la longitud del dato de la columna NIT
# para despues utilizar la funcion de contar valores para saber si hay NITs con uns longitud diferente a 9.

df1['NIT_LONG1']=[len(str(a)) for a in df1['nit_entidad']]
```

```python
df1['NIT_LONG1'].value_counts()
```

```python
df2['NIT_LONG2']=[len(str(a)) for a in df2['nit_entidad']]
```

```python
df2['NIT_LONG2'].value_counts()
```

De acuerdo con la información arrojada por la función aplicada, tenemos 7 diferentes longitudes en el campo de NIT del primer dataframe, en cambio para el segundo dataframe tenemos solamente 2 diferentes longitudes, por lo anterior, el proceso de limpieza será más exahustivo en el primer df.

```python
# Como en el df 2 tenemos valores con longitud de 9 y 11, vamos a limpiar las de 11 posiciones que son aquellas que traen el guion y el digito de
# verificación, así dejaremos todos los nits de este df2 de 9 digitos. Para ello utilizaremos la funcionalidad split a partir del simbolo '-', los que
# no tengan este simbolo se mantendrán iguales, para los que si lo tengan se traeran los valores antes del simbolo.
df2 ['nit_entidad_2']=[ a.split('-')[0] for a in df2["nit_entidad"]]
```

```python
# Volvemos a mostrar el df2 y validamos nuevamente la longitud del nuevo campo 'nit_entidad_2'
df2
```

```python
df2['NIT_LONG2']=[len(str(a)) for a in df2['nit_entidad_2']]
```

```python
df2['NIT_LONG2'].value_counts()
```

Como podemos observar ahora solo tenemos nits de 9 digitos y hemos dejado esta columna del df2 lista. 

```python
# Ahora vamos a limpiar la columna nit_entidad del df1, para ello reemplazaremos todos los caracteres que no sean digitos del 0 al 9 utilizando
# una expresion regular y el resultado lo almacenaremos en una nueva columna que llamaremos 'nit_entidad_limpio'

df1['nit_entidad_limpio'] = df1['nit_entidad'].str.replace('[^0-9]', '', regex=True)
```

```python
df1['nit_entidad_limpio']
```

```python
# Luego volvemos a calcular la longitud del campo limpio
df1['NIT_LONG1']=[len(str(a)) for a in df1['nit_entidad_limpio']]
```

```python
df1['NIT_LONG1'].value_counts()
```

Como vemos, pasamos de 7 diferentes longitudes a 5 diferentes longitudes. Para dejar el campo limpio eliminaremos los registros que no tienen datos validos en la columna nit, es decir que su longitud es igual a 0 o 1, tambien quitaremos los de longitud igual a 11 pues pueden tratarse de erores de digitación. Para los valores de longitud 10, quitaremos el ultimo digito que es posible que corresponda al digito de verificación.

```python
df1.drop(df1[(df1['NIT_LONG1']<9)].index, inplace=True)
```

```python
df1
```

Como vemos, el dataframe 1 ahora tiene menos filas, paso de tener 1000 a tener 952, pues se eliminaron los registros con nits de logitudes de 0 y 1. Ahora procederemos a borrar los de nits con longitudes mayores a 10.

```python
df1.drop(df1[(df1['NIT_LONG1']>10)].index, inplace=True)
```

```python
df1
```

Con lo anterior, se evidencia que fueron eliminados los registros con nits de una longitud de 11 pues pasamos de 952 filas a 872.

```python
df1 ['nit_entidad_limpio_9']=[ a[0:9] for a in df1["nit_entidad_limpio"]]
```

```python
df1['NIT_LONG1']=[len(str(a)) for a in df1['nit_entidad_limpio_9']]
```

```python
df1['NIT_LONG1'].value_counts()
```

Ya nuestro df1 quedo con 872 filas con nit valido, y ahora cumple con las mismas condiciones del df2. Ahora nombraremos las columnas de los NIT arreglados de la misma manera en cada uno de los dataframes.

```python
df2['nit_entidad_limpio_9']=df2['nit_entidad_2']
```

```python
#Vamos a renombrar las columnas que contienen los NITs limpios para que se llamen igual en ambos dataframes y poder realizar operaciones entre ambos
# conjuntos de datos.
df1 = df1.rename(columns={'nit_entidad_limpio_9': 'NIT_ENTIDAD_OK'})
```

```python
df2 = df2.rename(columns={'nit_entidad_limpio_9': 'NIT_ENTIDAD_OK'})
```

```python
df4=pd.merge(df1,df2,on='NIT_ENTIDAD_OK',how='inner')
```

```python
df4
```

Según el resultado, ninguno de los NITs del df1 se encuentra en el df2.

```python
#Probaremos con el nombre de entidad, estas columnas ya se llamaban igual
df6=pd.merge(df1,df2,on='nombre_entidad',how='inner')
```

```python
df6
```

Según el resultado, ninguno de los nombres de entidad del df1 se encuentra en el df2. Esto puede ser consecuencia de que algunos de los valores de las columnas de los dataframes fueron modificados (se observa que no tienen letras con tildes, por ejemplo, si el nombre es Andrés, en algunos casos se observa Andrs.

```python
# Para continuar probando las funciones para relacionar tablas, ahora utilizaremos un concat para unir los conjuntos de datos y luego buscaremos 
# registros duplicados por NIT (NIT duplicados).
df5=pd.concat([df1,df2])
```

```python
df7=pd.merge(df1,df5,on='NIT_ENTIDAD_OK',how='inner')
```

```python
duplicateRows = df7[df7.duplicated(['NIT_ENTIDAD_OK'])]
```

```python
duplicateRows
```

Se encontraron 169 registros de NITs duplicados en la unión de ambos conjuntos de datos.
