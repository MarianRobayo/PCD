3. Cruce de bases

```python
# Para poder trabajar con ambos conjuntos de datos y aprovechando que tienen las mismas columnas los uniremos, pero antes de ello le crearemos a c/u de
# los dfs un campo llamado 'class' que me permita identificar a que clase de vuelo corresponde cada registro.
df1['class']= "business"
```

```python
df1
```

```python
df2['class']= "economy"
```

```python
df2
```

```python
# Ahora uniremos ambos dataframes en uno solo al que llamaremos df3, le diremos que ignore el indice del df original y traiga uno nuevo para el nuevo
# conjunto de datos
df3=pd.concat([df1, df2],ignore_index=True)
```

```python
df3
```

```python
df3.info()
```

Al aplicar la funcion info al nuevo df (df3) podemos corroborar que se mantienen las mismas 12 columnas/variables y que el total de registros corresponde con la suma de la cantidad de registros del df1 más el df2. De igual forma los tipos de variables se mantienen. 

```python
# Al calcular los valores unicos de la columna class podemos ver la correspondencia entre el numero de registros de los df1 y df2.
df3['class'].value_counts()
```

```python
# Al realizar una exploracion visual de la variable stop nos percatamos que los valores allí registrados no coinciden con lo indicado en el diccionario
# de datos, por lo que es necesario realizar una limpieza a esta columna dejando unicamente los tres valores categoricos: non-stop, 1-stop y 2-stop, 
# primero revisaremos los valores unicos de esta variable para determinar el camino más adecuado en su limpieza.
df3['stop'].value_counts()
```

```python
# unicamente uno de los valores reporta el simbolo '+' por lo que procederemos a limpiarlo.
df3['stop'] = df3['stop'].str.replace('+','')
```

```python
df3['stop'].value_counts()
```

```python
# Ahora dejaremos los valores de la columna hasta 8 posiciones ya que es la máxima longitud en cuanto a las tres posibilidades de la variable
df3['stop']=df3['stop'].str.slice(stop=8)
```

```python
df3['stop'].value_counts()
```

```python
# De esta manera ya quedamos con nuestras 3 categorias, solo resta limpiar la categoria 1-stop quitando el excedente '\n\t'
df3['stop'] = df3['stop'].str.replace('\n\t','')
```

```python
df3['stop'].value_counts()
```

4. Filtro de bases

```python
# Ahora revisaremos las aerolineas que estan dentro de nuestro dataframe d3 para identificar cuales de estas ofrecen clase business y clase economy
df3['airline'].value_counts()
```

```python
# Revisando las aerolineas que tienen registros con clase business
df3[ df3['class'] == 'business']['airline'].value_counts()
```

```python
# Revisando las aerolineas que tienen registros con clase economy
df3[ df3['class'] == 'economy']['airline'].value_counts()
```

De acuerdo con los resultados, se observa que unicamente las aerolineas Vistara y Air India ofrecen vuelos en clase business, mientras que todas las aerolineas del df 3 tienen vuelos en clase economy.

5. Gráficas
   Primero transformaremos la variable price a valor numerico para poder utilizarla en los analisis graficos

```python
df3['price_num']=[float(str(a).replace(',','')) for a in df3['price']]
```

```python
df3.info()
```

```python
df3['price_num'].describe()
```

El precio más alto para un vuelo es de 123071 y el minimo es de 1105.

```python
# Antes de realizar las gráicas realizaremos unas tablas de contingencia en la cual podemos observar de manera ordenada las relaciones entre variables
# categoricas, como habiamos analizado previamente por separado en cuanto a las variables airline y class
pd.crosstab(index=df3['class'], columns=df3['airline'], margins=True)
```

```python
plot = pd.crosstab(index=df3['class'], columns=df3['airline']).apply(lambda r: r/r.sum() *100, axis=1).plot(kind='bar')
```

```python
plot=pd.crosstab(index=df3['class'], columns=df3['airline']).plot(kind='bar', stacked=True)
```

De acuerdo con las gráficas anteriores se tiene evidencia estadistica para decir que la aerolinea con mayor cantidad de vuelos operados tanto en clase economy como clase business es Vistara.

```python
# Crearemos unos diagrams de bigotes para analizar el comportamiento de los precios de acuerdo con la aerolinea y clase.
sns.boxplot(data=df3, x=df3['airline'], y=df3['price_num'])
```

```python
sns.boxplot(data=df3, x=df3['class'], y=df3['price_num'])
```

Teniendo en cuenta lo anterior, de la primera gráfica podemos inferir que los precios de vuelos ofrecidos por las aerolineas Air India, Vistara y StarAir no presentan datos atipicos, en las demás aerolineas si se presentan bastantes datos atipicos. 

Con respecto a la segunda gráfica podemos observar que ambas clases presentan datos atipicos con respecto a la variable precio, sin embargo, la clase economy no tiene valores atipicos por debajo de un desviación estandar sino por encima. En la clase business se tienen valores atipicos tanto por arriba y por debajo, siendo más frecuente por arriba de una desviación estandar.
